```python
import pandas as pd

def replace_outliers(df):
    for column in df.select_dtypes(include=['number']).columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lb = Q1 - 1.5 * IQR
        ub = Q3 + 1.5 * IQR
        
        def replace_value(x):
            if x < lb:
                return Q1
            elif x > ub:
                return Q3
            else:
                return x
        
        df[column] = df[column].apply(replace_value)
    
    return df
```

```python
df = replace_outliers(df)
```
